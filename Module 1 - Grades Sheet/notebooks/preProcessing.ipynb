```python
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

# Loading image
img = cv.imread("../data/1.jpg")
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Converting to grayscale
grayScale = cv.cvtColor(img_rgb, cv.COLOR_RGB2GRAY)
thresholded = grayScale.copy()

# Experiment 1: Trying global thresholding directly without applying gaussian
thresholded = np.where(grayScale > 127, 0, 255).astype(np.uint8)


plt.figure(figsize=(12,16))
plt.title("Global Threshold Only", fontsize=16)
plt.imshow(thresholded, cmap='gray')
plt.axis('off')
plt.show()

# Experiment 2: Apply Gaussian after Global Threshold
blur_after_global = cv.GaussianBlur(thresholded, (5, 5), 0)

plt.figure(figsize=(12,16))
plt.title("Gaussian Blur After Global Threshold", fontsize=16)
plt.imshow(blur_after_global, cmap='gray')
plt.axis('off')
plt.show()
```
